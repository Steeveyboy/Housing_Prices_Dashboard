# Transformation (made by Fabian Nwaoha)

```python
import pandas as pd
```

```python
df_weather = pd.read_csv('weather_stations.csv', parse_dates=['yearmo'])
df_weather
```

```python
#fixing date pattern
df_weather['date']=df_weather['yearmo'].str[:4]+'-'+df_weather['yearmo'].str[4:]


df_weather['date'] = pd.to_datetime(df_weather.date)
df_weather
```

```python
df_weather.dtypes
```

```python
#Selecting important features in data
df_weather = df_weather[['Stn_Name','Clim_ID','Prov_or_Ter','Tm','Tx','Tn','P','S_G','date']]
df_weather
```

```python
df_weather.isnull().sum()
```

```python
#dropping rows with ####### useless values
df_weather = df_weather[df_weather.P != '######']
df_weather['P'] = df_weather['P'].astype(float)
```

```python
#filing nan values with column averages
df_weather['S_G'].fillna((df_weather['S_G'].mean()), inplace=True)
df_weather['Tm'].fillna((df_weather['Tm'].mean()), inplace=True)
df_weather['Tx'].fillna((df_weather['Tx'].mean()), inplace=True)
df_weather['Tn'].fillna((df_weather['Tn'].mean()), inplace=True)
df_weather['P'].fillna((df_weather['P'].astype(float).mean()), inplace=True)
df_weather
```

```python
#Grouping by location and date to create regional averages
regional_mean=df_weather.groupby(['Prov_or_Ter','date'])[['Tm','Tx','Tn','P','S_G']].mean()
regional_mean=regional_mean.reset_index()
regional_mean.rename(columns={'Tm': 'Tm Average','Prov_or_Ter':'location', 'Tx': 'Tx Average', 'Tn': 'Tn Average',  'P': 'P Average',  'S_G': 'S_G Average'}, inplace=True)
regional_mean
```

```python
#Importing second dataset
housing = pd.read_csv('structure_housing_data_22-Mar-23_08-30PM.csv')
housing
```

```python
regional_mean.location.unique()
```

```python
#changing locations to acronyms to match other dataset
housing.loc[housing["location"] == "British Columbia", "location"] = "BC"
housing.loc[housing["location"] == "Ontario", "location"] = "ON"
housing.loc[housing["location"] == "Newfoundland and Labrador", "location"] = "NL"
housing.loc[housing["location"] == "Prince Edward Island", "location"] = "PE"
housing.loc[housing["location"] == "Nova Scotia", "location"] = "NS"
housing.loc[housing["location"] == "New Brunswick", "location"] = "NB"
housing.loc[housing["location"] == "Quebec", "location"] = "QC"
housing.loc[housing["location"] == "Manitoba", "location"] = "MB"
housing.loc[housing["location"] == "Saskatchewan", "location"] = "SK"
housing.loc[housing["location"] == "Alberta", "location"] = "AB"
housing.loc[housing["location"] == "British Columbia", "location"] = "BC"
housing.loc[housing["location"] == "British Columbia", "location"] = "BC"
housing.loc[housing["location"] == "British Columbia", "location"] = "BC"
housing.loc[housing["location"] == "British Columbia", "location"] = "BC"
```

```python
#standardizing date format
housing['date'] = pd.to_datetime(housing.date)
housing
```

```python
#merging 
new_result = pd.merge(regional_mean, housing, how="left", on=["location","date"])
```

```python
new_result.isnull().sum()
new_result
```

```python
#creating surrogate keys
new_result['Surrogate Keys'] = range(1,len(new_result)+1)
new_result = new_result.reindex(columns=['Surrogate Keys'] + list([c for c in new_result.columns if c!= 'Surrogate Keys']))
```

```python
fact_table = new_result[new_result['total'].notna()]
fact_table
#fact table
```

```python
location_dimension = fact_table[['location']]
location_dimension = location_dimension.drop_duplicates()
location_dimension
#Location dimension
```

```python
date_dimension = fact_table[['date']]
date_dimension['Month'] = date_dimension['date'].dt.month 
date_dimension['Year'] = date_dimension['date'].dt.year

date_dimension = date_dimension.drop(['date'], axis=1)

date_dimension
#date dimension
```

```python
# create a table 
def create_table(connection):
    cursor = connection.cursor()
    cursor.execute('''CREATE TABLE IF NOT EXISTS "4142_table" (
                          "Surrogate Keys" INTEGER PRIMARY KEY,
                          location TEXT,
                          date TIMESTAMP,
                          "Tm Average" FLOAT,
                          "Tx Average" FLOAT,
                          "Tn Average" FLOAT,
                          "P Average" FLOAT,
                          "S_G Average" FLOAT,
                          total FLOAT,
                          house FLOAT,
                          land FLOAT);''')
    connection.commit()
    cursor.close()
    
create_table(conn)
conn.close()
```

```python
# load the df into the table
from sqlalchemy import create_engine

def load_dataframe(dataframe, table_name, connection):
    engine = create_engine(f"postgresql+psycopg2://{connection.info.user}:{connection.info.password}@{connection.info.host}:{connection.info.port}/{connection.info.dbname}")
    dataframe.to_sql(table_name, engine, if_exists='replace', index=False, schema='public', method='multi')


conn = create_conn()
load_dataframe(fact_table, '4142_table', conn)
conn.close()
```

```python
# validate the df is loaded successfully
def show_table(table,connection):
    query = f"SELECT * FROM \"{table}\""
    dataframe = pd.read_sql_query(query, connection)
    connection.close()
    return dataframe

conn = create_conn()
table_df = show_table('4142_table',conn)
conn.close()
```
